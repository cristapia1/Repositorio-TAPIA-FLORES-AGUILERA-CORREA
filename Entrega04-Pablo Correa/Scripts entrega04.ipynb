```python
# Cargar el archivo Excel desde Google Colab
data = pd.read_excel('/content/base2.xlsx', header=0)
```

```python
# Imprimir los nombres de las columnas
print(data.columns)

# Imprimir las primeras filas del DataFrame
print(data.head())
```

```python
# Eliminar las filas correspondientes a "Total País", "Región Metropolitana" y "Provincia de Santiago"
data_filtered = data[~data['Unidad Territorial'].isin(['TOTAL PAIS', 'Region Metropolitana', 'Provincia de Santiago'])]

# Crear la visualización de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura según sea necesario
sns.barplot(x='2021', y='Unidad Territorial', data=data_filtered)
plt.title('Cantidad de delitos por comuna en 2021 (sin Total País, Región Metropolitana ni Provincia de Santiago)')
plt.xlabel('Cantidad de delitos')
plt.ylabel('Comuna')
plt.show()
```
